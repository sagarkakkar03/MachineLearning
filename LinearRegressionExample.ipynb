```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Salary_Data.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['YearsExperience'].fillna(df['YearsExperience'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
```

```python
features = df.drop('Salary', axis=1)
label = df['Salary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, label, random_state=22, test_size=.2)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
for rs in range(1, 101):
  X_train, X_test, y_train, y_test = train_test_split(features, label, random_state=rs, test_size=.2)
  model = LinearRegression()
  model.fit(X_train, y_train)
  print(model.score(X_test, y_test))
```
