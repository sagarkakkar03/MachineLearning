```python
df.drop(columns = ['Suburb', 'Address', 'Method', 'Date', 'SellerG', 'CouncilArea', 'Lattitude', 'Longtitude', 'BuildingArea', 'YearBuilt'], inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False)
ohe_Type = ohe.fit_transform(df[['Type']])
```

```python
type_column_names = ohe.get_feature_names_out(['Type'])
df_type_encoded = pd.DataFrame(ohe_Type, columns=type_column_names, index=df.index)
df_type_encoded
```

```python
df = pd.concat([df.drop('Type', axis=1), df_type_encoded], axis=1)
print(df.head())
```

```python
ohe_Region = ohe.fit_transform(df[['Regionname']])
```

```python
region_column_names = ohe.get_feature_names_out(['Regionname'])
df_region_encoded = pd.DataFrame(ohe_Region, columns=region_column_names, index=df.index)
df_region_encoded
```

```python
df['Car'].fillna(1, inplace=True)
```

```python
df.isnull().sum()
```

```python
final_df = pd.concat([df.drop(['Regionname'], axis=1), df_region_encoded], axis = 1)
```
